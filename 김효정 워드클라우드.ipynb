```python
import requests
import pandas as pd
import lxml.html
```

```python
res=requests.get('https://www.nytimes.com/2020/04/24/world/europe/disinformation-china-eu-coronavirus.html')
```

```python
root=lxml.html.fromstring(res.text)
```

```python
content=root.cssselect('[itemprop="articleBody"]')[0].text_content()
```

```python
content
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(stop_words='english')
```

```python
tdm=cv.fit_transform([content]).toarray()
```

```python
df=pd.DataFrame({'words':cv.get_feature_names(), 'freq':tdm.flat})
```

```python
df.sort_values('freq', ascending=False)
```

```python
from wordcloud import WordCloud
```

```python
wc=WordCloud(width=500,height=400,background_color='white')
```

```python
frequency={word:freq for word,freq in zip(cv.get_feature_names(),tdm.flat)}
```

```python
wc.fit_words(frequency)
```

```python
wc.to_image()
```
